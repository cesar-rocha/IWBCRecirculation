```python
import numpy as np
import cmocean
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import seawater as sw
```

```python
mean = np.load('data/vel_twomonth_mean.npz')
snap = np.load('data/vel_snapshot_daily-averaged.npz')
lon,lat = mean['lon'],mean['lat']
lon,lat = np.meshgrid(lon,lat)
```

```python
m = Basemap(projection='merc',llcrnrlat=-27.,urcrnrlat=-18.,\
            llcrnrlon=-45.,urcrnrlon=-38.,lat_ts=-23.,resolution='i')
```

```python
x,y = m(lon,lat)
```

## Two-month average (March-April 2011)

```python
fig = plt.figure(figsize = (14,8))

plt.subplot(121)
m.drawcoastlines()
m.etopo()
m.fillcontinents(color='0.5',lake_color=None)

dec = 4
Q = m.quiver(x[::dec,::dec],y[::dec,::dec],mean['u'][0,::dec,::dec],mean['v'][0,::dec,::dec],linewidth=.5,scale=7)
qk = plt.quiverkey(Q, 0.5, 0.92, 1, r'$1$ m s$^{-1}$', labelpos='W',
                   fontproperties={'weight': 'bold'})

m.drawparallels(np.arange(-90.,91.,5.),labels=[1,0,0,0],fontsize=14)
m.drawmeridians(np.arange(-180.,181.,5.),labels=[0,0,0,1],fontsize=14)

plt.title('0 m')

plt.subplot(122)
m.drawcoastlines()
m.etopo()
m.fillcontinents(color='0.5',lake_color=None)

dec = 4
Q = m.quiver(x[::dec,::dec],y[::dec,::dec],mean['u'][51,::dec,::dec],mean['v'][51,::dec,::dec],linewidth=.5,scale=7)
qk = plt.quiverkey(Q, 0.5, 0.92, 1, r'$1$ m s$^{-1}$', labelpos='W',
                   fontproperties={'weight': 'bold'})

m.drawparallels(np.arange(-90.,91.,5.),labels=[1,0,0,0],fontsize=14)
m.drawmeridians(np.arange(-180.,181.,5.),labels=[0,0,0,1],fontsize=14)

plt.title('1000 m')
```

```python
cv = np.arange(-.3,.35,.05)

fig = plt.figure(figsize = (14,16))

plt.subplot(221)

plt.contourf(mean['lon'],-mean['z'],mean['u'][:,200],cv,extend='both')
plt.contour(mean['lon'],-mean['z'],mean['u'][:,200],cv,colors='k')
plt.ylim(2500,0)
plt.xlim(-45,-38)
plt.ylabel('Depth [m]')
plt.xticks([])

plt.title(r'u, 21$^\circ$S')

plt.subplot(222)

plt.contourf(mean['lon'],-mean['z'],mean['v'][:,200],cv,extend='both')
plt.contour(mean['lon'],-mean['z'],mean['v'][:,200],cv,colors='k')
plt.yticks([])
plt.ylim(2500,0)
plt.xlim(-45,-38)
plt.xticks([])
plt.title(r'u, 21$^\circ$S')


plt.subplot(223)

plt.contourf(mean['lon'],-mean['z'],mean['u'][:,80],cv,extend='both')
plt.contour(mean['lon'],-mean['z'],mean['u'][:,80],cv,colors='k')
plt.ylim(2500,0)
plt.xlim(-45,-38)
plt.ylabel('Depth [m]')
plt.xlabel('Longitude')
plt.title(r'u, 25$^\circ$S')

plt.subplot(224)
plt.title(r'v, 25$^\circ$S')

cb = plt.contourf(mean['lon'],-mean['z'],mean['v'][:,80],cv,extend='both')
plt.contour(mean['lon'],-mean['z'],mean['v'][:,80],cv,colors='k')
plt.yticks([])
plt.ylim(2500,0)
plt.xlim(-45,-38)
plt.xlabel('Longitude')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.2, 0.02, 0.6])
fig.colorbar(cb, cax=cbar_ax,label=r'Velocity [m s$^{-1}]$')

```
